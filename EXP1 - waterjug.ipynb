```python
#BFS
from collections import deque

def water_jug(j1,j2,target):
    q= deque([((0,0),[])])
    visited = set()

    while q:
        (x,y),path = q.popleft()
        if(x,y) in visited:
            continue;
        visited.add((x,y))
        path=path+[(x,y)]

        if x==target or y==target:
            return path
        for nx,ny in [
            (j1,y),(x,j2),#fill
            (0,y),(x,0), #empty
            (x- min(x,j2-y),y+min(x,j2-y)), #1->2
            (x+ min(y,j1-x),y-min(y,j1-x)) #2->1
        ]:
            if(nx,ny) not in visited:
                q.append(((nx,ny),path))
    return None

j1 = int(input("enter the capacity of jug 1: "))
j2 = int(input("enter the capacity of jug 2: "))
target = int(input("enter target amount: "))
solution = water_jug(j1,j2,target)
if(solution):
    print("solution path :")
    for step in solution:
        print(step)
else:
    print("No solution possible")
```

```python
#DFS
def water_jug_dfs(j1, j2, target, x=0, y=0, path=None, depth=0, max_depth=20):
    if path is None:
        path = []

    if depth > max_depth:
        return None

    # avoid repeating the same state in current path
    if (x, y) in path:
        return None

    path.append((x, y))

    if x == target or y == target:
        return path

    moves = [
        (j1, y),  # fill jug1
        (x, j2),  # fill jug2
        (0, y),   # empty jug1
        (x, 0),   # empty jug2
        (x - min(x, j2 - y), y + min(x, j2 - y)),  # jug1 -> jug2
        (x + min(y, j1 - x), y - min(y, j1 - x))   # jug2 -> jug1
    ]

    for nx, ny in moves:
        # skip useless move (no state change)
        if nx == x and ny == y:
            continue

        result = water_jug_dfs(j1, j2, target, nx, ny, path.copy(), depth + 1)
        if result:
            return result

    return None

j1 = int(input("enter the capacity of jug 1: "))
j2 = int(input("enter the capacity of jug 2: "))
target = int(input("enter target amount: "))

solution = water_jug_dfs(j1, j2, target)

if solution:
    print("Solution path:")
    for step in solution:
        print(step)
else:
    print("No solution possible")
```

```python
#Memoization
def water_jug_dfs_memo(j1, j2, target):
    visited = set()

    def dfs(x, y, path):
        if (x, y) in visited:
            return None

        visited.add((x, y))
        path.append((x, y))

        if x == target or y == target:
            return path

        moves = [
            (j1, y),  # fill jug1
            (x, j2),  # fill jug2
            (0, y),   # empty jug1
            (x, 0),   # empty jug2
            (x - min(x, j2 - y), y + min(x, j2 - y)),  # jug1 -> jug2
            (x + min(y, j1 - x), y - min(y, j1 - x))   # jug2 -> jug1
        ]

        for nx, ny in moves:
            if (nx, ny) not in visited:
                result = dfs(nx, ny, path.copy())
                if result:
                    return result

        return None

    return dfs(0, 0, [])
j1 = int(input("enter the capacity of jug 1: "))
j2 = int(input("enter the capacity of jug 2: "))
target = int(input("enter target amount: "))

solution = water_jug_dfs_memo(j1, j2, target)

if solution:
    print("Solution path:")
    for step in solution:
        print(step)
else:
    print("No solution possible")
```
