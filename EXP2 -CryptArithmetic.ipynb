```python
def crypt_arithmetic_csp(words, result):

    letters = list(set("".join(words) + result))
    if len(letters) > 10:
        print("Invalid input - more than 10 letters")
        return

    first_letters = set(word[0] for word in words + [result])
    assignment = {}
    used_digits = set()

    def word_value(word):
        return int("".join(str(assignment[c]) for c in word))

    def backtrack(index):
        if index == len(letters):
            return sum(word_value(w) for w in words) == word_value(result)

        letter = letters[index]

        for digit in range(10):
            if digit in used_digits:
                continue
            if digit == 0 and letter in first_letters:
                continue

            assignment[letter] = digit
            used_digits.add(digit)

            if backtrack(index + 1):
                return True

            del assignment[letter]
            used_digits.remove(digit)

        return False

    if backtrack(0):
        print("Solution Found:")
        for k in sorted(assignment):
            print(f"{k} = {assignment[k]}")
    else:
        print("No Solution")



expr = input("Enter crypt arithmetic (eg: SEND+MORE=MONEY): ").replace(" ", "")
left, right = expr.split("=")
words = left.split("+")

crypt_arithmetic_csp(words, right)
```
