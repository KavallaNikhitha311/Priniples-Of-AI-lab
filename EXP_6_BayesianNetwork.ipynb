```python
import pandas as pd
#from pgmpy.models import BayesianNetwork
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination

# Step 1: Create dataset
data = pd.DataFrame({
    'Rain': ['Yes', 'No', 'Yes', 'Yes', 'No'],
    'Sprinkler': ['On', 'Off', 'Off', 'On', 'Off'],
    'WetGrass': ['Yes', 'No', 'Yes', 'Yes', 'No']
})

# Step 2: Define structure
model = DiscreteBayesianNetwork([
    ('Rain', 'WetGrass'),
    ('Sprinkler', 'WetGrass')
])

# Step 3: Learn probabilities
model.fit(data, estimator=MaximumLikelihoodEstimator)

# Step 4: Inference
infer = VariableElimination(model)

# Query example
result = infer.query(variables=['Rain'], evidence={'WetGrass': 'Yes'})
print(result)
```

```python
import pandas as pd
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination

data = pd.DataFrame({
    'StudyTime': ['High', 'Low', 'High', 'Low', 'High', 'Low', 'High', 'Low','Low' ],
    'Attendance': ['Good', 'Good', 'Poor', 'Poor', 'Good', 'Poor', 'Good', 'Good', 'Good'],
    'ExamResult': ['Pass', 'Fail', 'Pass', 'Fail', 'Pass', 'Fail', 'Pass', 'Fail', 'Fail']
})

model = DiscreteBayesianNetwork([
    ('StudyTime', 'ExamResult'),
    ('Attendance','ExamResult')
])

model.fit(data, estimator=MaximumLikelihoodEstimator)
infer=VariableElimination(model)
result=infer.query(variables=['ExamResult'],evidence={'Attendance':'Good'})
print(result)
```

```python
#!pip install pgmpy
import pandas as pd
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination

data = pd.DataFrame({
    'IncomeStability': ['Stable', 'Unstable', 'Stable', 'Stable', 'Unstable', 'Stable', 'Unstable', 'Stable'],
    'CreditHistory': ['Good', 'Bad', 'Good', 'Bad', 'Bad', 'Good', 'Bad', 'Good'],
    'EmploymentType': ['Salaried', 'Self', 'Self', 'Salaried', 'Self', 'Salaried', 'Self', 'Salaried'],
    'DefaultRisk': ['No', 'Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No']
})

model = DiscreteBayesianNetwork([
    ('IncomeStability', 'DefaultRisk'),
    ('CreditHistory', 'DefaultRisk'),
    ('EmploymentType', 'DefaultRisk')
])

model.fit(data, estimator=MaximumLikelihoodEstimator)
infer = VariableElimination(model)
# Query 1: Default probability given bad credit
q1 = infer.query(variables=['DefaultRisk'], evidence={'CreditHistory': 'Bad'})
print("\n Evidence: CreditHistry ")
print(q1)
# Query 2: Default probability with incomplete data
q2 = infer.query(variables=['DefaultRisk'], evidence={'IncomeStability': 'Unstable'})
print("\n  Evidence: IncomeStability ")
print(q2)
# Query 3: Combined evidence
q3 = infer.query(
    variables=['DefaultRisk'],
    evidence={'CreditHistory': 'Bad', 'EmploymentType': 'Self'}
)
print(" \n Evidence: CreditHistory & EmploymentType ")
print(q3)
```
